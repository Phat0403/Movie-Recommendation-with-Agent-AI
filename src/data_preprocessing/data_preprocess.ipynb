### Name Basics

- nconst (string) - định danh duy nhất alphanumeric của tên/người
- primaryName (string) - tên mà người đó thường được ghi nhận nhất
- birthYear - theo định dạng YYYY
- deathYear - theo định dạng YYYY nếu có, nếu không thì là '\N'
- primaryProfession (mảng string) - 3 nghề nghiệp hàng đầu của người đó
- knownForTitles (mảng tconsts) - các tiêu đề mà người đó được biết đến

```python
name_basics = pd.read_csv('./data/name.basics.tsv', sep='\t')
name_basics
```

```python
name_basics.info()
```

```python
name_basics.replace("\\N", np.nan, inplace=True)
```

```python
name_basics
```

```python
name_basics['birthYear'] = pd.to_numeric(name_basics['birthYear'], errors='coerce')
name_basics['deathYear'] = pd.to_numeric(name_basics['deathYear'], errors='coerce')
```

```python
name_basics
```

```python
name_basics.info()
```

```python
name_basics.to_csv('./data/name_basics.csv', index=False)
```

### Title Basics

- tconst (string) - định danh duy nhất của tiêu đề, bao gồm chữ và số
- titleType (string) – loại/định dạng của tiêu đề (ví dụ: phim, ngắn, series truyền hình, tập truyền hình, video, v.v.)
- primaryTitle (string) – tiêu đề phổ biến hơn / tiêu đề được sử dụng bởi các nhà làm phim trong tài liệu quảng cáo tại thời điểm phát hành
- originalTitle (string) - tiêu đề gốc, bằng ngôn ngữ gốc
- isAdult (boolean) - 0: tiêu đề không dành cho người lớn; 1: tiêu đề dành cho người lớn
- startYear (YYYY) – đại diện cho năm phát hành của một tiêu đề. Trong trường hợp series truyền hình, đây là năm bắt đầu của series
- endYear (YYYY) – năm kết thúc của series truyền hình. '\N' cho tất cả các loại tiêu đề khác
- runtimeMinutes – thời gian chạy chính của tiêu đề, tính bằng phút
- genres (mảng string) – bao gồm tối đa ba thể loại liên quan đến tiêu đề

```python
title_basics = pd.read_csv('./data/title.basics.tsv', sep='\t')
title_basics
```

```python
title_basics.info()
```

```python
title_basics.replace("\\N", np.nan, inplace=True)
title_basics['startYear'] = pd.to_numeric(title_basics['startYear'], errors='coerce')
title_basics['runtimeMinutes'] = pd.to_numeric(title_basics['runtimeMinutes'], errors='coerce')
title_basics.info()
```

```python
title_basics
```

```python
title_basics['endYear'].isnull().sum()
```

```python
title_basics.to_csv('./data/title_basics.csv', index=False)
```

### Title akas

- titleId (chuỗi) - một tconst, là định danh duy nhất alphanumeric của tiêu đề
- ordering (số nguyên) – một số để xác định duy nhất các hàng cho một titleId nhất định
- title (chuỗi) – tiêu đề đã được địa phương hóa
- region (chuỗi) - khu vực cho phiên bản này của tiêu đề
- language (chuỗi) - ngôn ngữ của tiêu đề
- types (mảng) - Tập hợp các thuộc tính được liệt kê cho tiêu đề thay thế này. Một hoặc nhiều trong số các giá trị sau: "alternative", "dvd", "festival", "tv", "video", "working", "original", "imdbDisplay". Các giá trị mới có thể được thêm vào trong tương lai mà không cần thông báo
- attributes (mảng) - Các thuật ngữ bổ sung để mô tả tiêu đề thay thế này, không được liệt kê
- isOriginalTitle (boolean) – 0: không phải tiêu đề gốc; 1: tiêu đề gốc

### Title Episode

- tconst (string) - định danh duy nhất alphanumeric của tập phim
- parentTconst (string) - định danh duy nhất alphanumeric của loạt phim cha
- seasonNumber (integer) – số mùa mà tập phim thuộc về
- episodeNumber (integer) – số tập của tconst trong loạt phim

```python
title_episosde = pd.read_csv('./data/title.episode.tsv', sep='\t')
title_episosde
```

```python
title_episosde.info()
```

```python
title_episosde.replace('\\N', np.nan, inplace=True)
title_episosde['seasonNumber'].fillna(-1, inplace=True)
title_episosde['episodeNumber'].fillna(-1, inplace=True)
title_episosde['seasonNumber'] = title_episosde['seasonNumber'].astype(int)
title_episosde['episodeNumber'] = title_episosde['episodeNumber'].astype(int)
title_episosde.info()
```

```python
title_episosde.to_csv('./data/title_episodes.csv', index=False)
```
